```python
# We configurate the font and palette, using the "inferno palette"
sns.set_palette("inferno")


# Create the figure and axis
fig, ax = plt.subplots()


# Set the figure and axis background color to black
fig.patch.set_facecolor('black')  # Set the figure's background color to black
ax.set_facecolor('black')         # Set the chart area's background color to black


# Using 'sns.hisplot()' , we create the chart with Seaborn.
sns.histplot(x = "age", 
             data = df_onlinefoods, 
             edgecolor = 'fuchsia',
             color = "violet",
             ax = ax,
             kde = True, 
             bins = 20 )


# Using 'ax.set_xlabel/ylabel()', we name the axis and edit the colour and size of the labels.
ax.set_xlabel("Age", fontsize=16, color='white')  # Changes the name, colour and font size of axis X
ax.set_ylabel("Count", fontsize=16, color='white') # Changes the name, colour and font size of axis Y


# Using 'ax.spines[]', we edit the axis.
ax.spines['top'].set_visible(False)     # Hides the top line
ax.spines['right'].set_visible(False)   # Hides the right line
ax.spines['left'].set_linewidth(2.5)    # Makes the Y axis line thicker
ax.spines['bottom'].set_linewidth(2.5)  # Makes the X axis line thicker
ax.spines['left'].set_color('yellow')   # Changes the colour of the Y axis
ax.spines['bottom'].set_color('yellow') # Changes the colour of the X axis


# Using 'ax.tick_params()', we edit the colour of the numbers on the axis.
ax.tick_params(axis='x', colors='yellow')  # Edits the colour of the numbers of the X axis
ax.tick_params(axis='y', colors='yellow')  # Edits the colour of the numbers of the Y axis


# Using 'plt.title()', we name the chart.
plt.title("Distribution of Age in dataset.", fontsize=14, color='white', pad=30);
```

#### **<span style="color:#01bef4">Countplot</span> üî¢ - Useful for identifying the frequency of occurrences and comparing the distribution of categories within the Dataset.**

```python
# Create the countplot with a specific color palette
sns.set_palette("inferno")  


# Create the figure and axis to have more control over the plot's layout and properties
fig, ax = plt.subplots()


# Set the figure and axis background color to black
fig.patch.set_facecolor('black')  # Set the figure's background color to black
ax.set_facecolor('black')         # Set the chart area's background color to black


# Create the countplot on the given axis
sns.countplot(x = "gender", 
              data = df_onlinefoods,
              edgecolor = 'yellow',
              color = '#f9e68b',
              ax = ax)


# Customize the axis appearance
ax.tick_params(colors='fuchsia')       # Set the tick colors to white for visibility

ax.spines['top'].set_linewidth(2)    # Increase the top axis thickness
ax.spines['right'].set_linewidth(2)  # Increase the right axis thickness
ax.spines['left'].set_linewidth(2)   # Increase the left axis thickness
ax.spines['bottom'].set_linewidth(2) # Increase the bottom axis thickness
ax.spines['left'].set_color('fuchsia')  # Set the color of the spines to white
ax.spines['bottom'].set_color('fuchsia')  # Set the color of the spines to white

# Customize axis labels
ax.set_xlabel("Gender", fontsize=12, color='white')  # Set x-axis label color and size
ax.set_ylabel("Count", fontsize=12, color='white')   # Set y-axis label color and size


# Rotate x-axis labels for better readability
plt.xticks(rotation=90, color="white")  # Make x-tick labels white


# Show the plot
plt.show();
```
